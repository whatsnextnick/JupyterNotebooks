````python
import numpy as np
import matplotlib as mp


#Solving Linear Systems
# 1. **Task**: Consider the system of linear equations:
#    ```
#    3x + 2y = 11
#    x - y = 2
#    ```
#    Represent this system in matrix form and use NumPy to solve for `x` and `y`.

# Coefficient matrix A
A = np.array([[3, 2],
              [1, -1]])

# Constants matrix B
B = np.array([11, 2]) 

X = np.linalg.solve(A, B)

x, y = X
print(f"x = {x}")
print(f"y = {y}")
````

````python
# Eigenvalues and Eigenvectors
# 1. **Task**: Create a 2x2 matrix with values of your choice. Compute the following:
#    - Eigenvalues (use `numpy.linalg.eig`)
#    - Eigenvectors

# Creating a 2x2 matrix
A = np.array([[4, 2],
              [3, 1]])

# Calculating eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Matrix A:")
print(A)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
````

````python
# 2. **Bonus**: Visualize the matrix's action (transformation) on its eigenvectors.
import numpy as np
import matplotlib.pyplot as plt

# Creating a 2x2 matrix
A = np.array([[4, 2],
              [3, 1]])

# Calculating eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Creating a plot
fig, ax = plt.subplots(figsize=(5,5))

# Plotting eigenvectors
for i in range(2):
    plt.quiver(0, 0, eigenvectors[0, i], eigenvectors[1, i], angles='xy', scale_units='xy', scale=1, color=['r', 'b'])

# Applying matrix A to eigenvectors and plotting
for i in range(2):
    transformed_vector = np.dot(A, eigenvectors[:, i])
    plt.quiver(0, 0, transformed_vector[0], transformed_vector[1], angles='xy', scale_units='xy', scale=1, color=['g', 'y'], alpha=0.5)

# Setting plot limits and labels
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title("Action of matrix A on its eigenvectors")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Eigenvector 1", "Eigenvector 2", "A * Eigenvector 1", "A * Eigenvector 2"])

# Displaying the plot
plt.show()

````
