```python

a_random_1to10 = np.random.randint(0,10, (1,5)) 
a_random_1to10
```

```python
b_random_1to10 = np.random.randint(0,10, (1,5))
b_random_1to10
```

```python
addt_results = a_random_1to10 + b_random_1to10
mult_results = a_random_1to10 * b_random_1to10

print("The element-wise addition results are:", addt_results)
print("The element-wise multiplication results are:", mult_results)
```

```python
#Flatten the matrices to 1D arrays and then use np.dot()
dot_product = np.dot(a_random_1to10.flatten(), b_random_1to10.flatten())

print("The dot-product of the 2 arrays is:", dot_product)
```

```python
#Angle between the 2 vectors
a_magn = np.linalg.norm(a_random_1to10)
b_magn = np.linalg.norm(b_random_1to10)
cos_angle = dot_product / (a_magn * b_magn)
angle_rad = np.arccos(np.clip(cos_angle, -1.0, 1.0))

print(np.degrees(angle_rad), angle_rad)

```

```python
#BONUS: Vizualizing 2 vectors and their resultant from addition using matplotlib
import matplotlib.pyplot as plt

#setting up 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax
```

```python
#Redefining the vectors to visualize the first 3 dimensions of the vectors
a_random_1to10 = np.array([4, 4, 1])  # Taking the first three elements
b_random_1to10 = np.array([5, 9, 1])  # Taking the first three elements

#Compute the resultant vector of a-random_1to10 & b_random_1to10
resultant = a_random_1to10 + b_random_1to10

#Plot the Vectors 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the original vectors
ax.quiver(0, 0, 0, a_random_1to10[0], a_random_1to10[1], a_random_1to10[2], color='b', label='Vector A')
ax.quiver(0, 0, 0, b_random_1to10[0], b_random_1to10[1], b_random_1to10[2], color='r', label='Vector B')

# Plotting the resultant vector
ax.quiver(0, 0, 0, resultant[0], resultant[1], resultant[2], color='g', label='Resultant')

# Setting the limits for the plot
ax.set_xlim([0, 15])
ax.set_ylim([0, 15])
ax.set_zlim([0, 15])

# Setting labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Adding a legend
ax.legend()

# Display the plot
plt.show()
```

```python
### Exercise 2: Matrix Manipulations
# Task: Create a 3x3 matrix with random integers between 1 and 10. Perform the following operations:

# Compute the transpose of the matrix
# Compute the determinant (use numpy.linalg.det)
# Compute the inverse (use numpy.linalg.inv)

matrix = np.random.randint(1,10,(3,3))

print(matrix)
```
